```python
import pandas as pd
import numpy as np


data = pd.read_csv('uncleaned_netflix_titles_dataset.csv', encoding='latin1')
data.head(5)
data.shape # Rows & Columns
data.columns #Column names
data.info() # Data type + nulls
data.isnull().sum() # Missing values count
```

```python
data['director'].fillna("Not Specified", inplace=True)
data['cast'].fillna("Not Specified", inplace=True)
data['country'].fillna("Unknown", inplace=True)
data['rating'].fillna("Not Rated", inplace=True)
```

```python
data.isnull().sum()
```

```python
# Convert Date Column Properly
# Convert date_added to datetime:

data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
data['year_added'] = data['date_added'].dt.year
data.head(2)
```

```python
# Clean Duration Column (Important)
data[['duration_value', 'duration_type']] = data['duration'].str.split(' ', expand= True)

# Convert duration_value to numeric:
data['duration_value'] = pd.to_numeric(data['duration_value'], errors='coerce')

```

```python
# Remove Duplicates

data = data.drop_duplicates()
```

```python
# Strip Extra Spaces (Remove extra spaces in the column)
data.columns = data.columns.str.strip()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=['year_added', 'duration_value']).copy()
data['year_added'] = data['year_added'].astype(int)
data['duration_value'] = data['duration_value'].astype(int)
```

```python
data.duplicated().sum()
```

```python
data['duration_value'].head()
```

```python
data = data.dropna(subset=['date_added'])
data = data.dropna(subset= ['duration'])
data.isnull().sum()
```

Data Cleaning is done!

Next EDA

```python
data['type'].value_counts()
```

```python
# Are TV Shows increasing in recent years?

data.groupby(['year_added', 'type']).size().unstack()
```

```python
# Which country produces the most content?

data[data['country'] != 'Unknown']['country'] \
    .str.split(', ').explode() \
    .value_counts().head(10)
```

```python
# Which country produces more TV Shows vs Movies?

data.groupby(['country', 'type']).size().unstack().sort_values(by="Movie", ascending= False)
```

```python
# Your table shows many combined countries like:
# United States, New Zealand, Japan
# United States, Poland
# This means some content is co-produced.
# For deeper accuracy, you would explode countries before grouping.

country_type = data[data['country'] != "Unknown"]
country_type = country_type.assign(
    country = country_type['country'].str.split(', ')
).explode('country')

country_type.groupby(['country','type']).size().unstack().sort_values(by='Movie', ascending=False).head(10)
```

```python
# Let’s analyze ratings strategy.

data['rating'].value_counts()
```

```python
# Now let’s combine type + rating.
# This will show:
# Are Movies more mature?
# Are TV Shows more teen-focused?

data.groupby(['type', 'rating']).size().unstack()
```

```python
# Most Popular Genres
genres = data['listed_in'].str.split(', ').explode()
genres.value_counts().head(10)
```

```python
# Average Movie Duration
# Now let’s analyze movie length.

movies = data[data['type'] == 'Movie']
movies['duration_value'].describe()
```

```python
# KPI KPI (Key Performance Indicator)
# A measurable metric that shows business performance.

# KPI 1 — Total Content
total_titles = len(data)
total_titles

# KPI 2 — Total Movies
total_movies = data[data['type'] == 'Movie'].shape[0]
total_movies

# KPI 3 — Total TV Shows
total_tv = data[data['type'] == 'TV Show'].shape[0]
total_tv

# KPI 4 — % Movies vs TV Shows
movie_percent = (total_movies / total_titles) * 100
tv_percent = (total_tv / total_titles) * 100

movie_percent, tv_percent

# KPI 5 — Top Producing Country
data[data['country'] != 'Unknown']['country'].value_counts().idxmax()

# KPI 6 — Top Genre
genres = data['listed_in'].str.split(', ').explode()
genres.value_counts().idxmax()


# KPI 7 — Dominant Rating
data['rating'].value_counts().idxmax()


# KPI 8 — Average Movie Duration
movies['duration_value'].mean()
```
